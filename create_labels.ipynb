## Labeling Jupyter Notebook
The task in this notebook is to simply to create a csv file of the audio file names along with their labels.

```python
import pandas as pd
import glob
```

```python
files = glob.glob(r"D:\Kaggle\datasets\emotion_small\val/*.wav")
```

```python
files[0]
```

```python
file_name = []
happy_labels = [0]*len(files)
sad_labels = [0]*len(files)
disgust_labels = [0]*len(files)
fear_labels = [0]*len(files)
neutral_labels = [0]*len(files)
hap = 0
sad = 0
neu = 0
```

```python
cnt = 0
for file in files:
    fname = file.replace("D:\\Kaggle\\datasets\\emotion_small\\val\\","")
    file_name.append(fname)
    if(fname.find("HAP")>-1 ):
        happy_labels[cnt] = 1
        hap = hap + 1
    elif(fname.find("NEU")>-1 ):
        neutral_labels[cnt] = 1
        neu = neu + 1
    elif(fname.find("SAD")>-1 ):
        sad_labels[cnt] = 1
        sad = sad + 1
    elif(fname.find("DIS")>-1):
        disgust_labels[cnt] = 1
    elif(fname.find("FEA")>-1):
        fear_labels[cnt] = 1
    cnt = cnt + 1
```

```python
df = pd.DataFrame({"fname":file_name,"happy":happy_labels,"sad":sad_labels,"disgust":disgust_labels,"fear":fear_labels,"neutral":neutral_labels})
```

```python
df
```

```python
cd = pd.DataFrame() 
for i in range(len(df)):
    if(len(set(df.iloc[i,1:])) == 2):
        cd = pd.concat((cd,df.loc[i:i]))
print(len(cd))
cd = cd.reset_index(drop = True)
```

```python
cd
```

```python
cd.to_csv(r"D:\Kaggle\datasets\emotion_small\meld_val_small.csv",index = False,header = True)
```
